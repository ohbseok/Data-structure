```python
# 파이썬의 리스트로 스택 구현
top = []
def isEmpty() :
  return len(top) == 0
  # True/False로 반환하려면 return 해주기 
def push(item) :
  top.append(item)
## pop : 삭제 
def pop() :
  if not len(top) == 0 :    # isEmpty 선언했으니 써도 됨
    return top.pop(-1)

def peek() :
  if not len(top) == 0 :
    return top[-1] 
    # pop으로 하나 꺼내서 확인하고 다시 스택에 push하라며
def size() :
  return len(top)
def clear() :
  # 여기서 전역변수로 지정
  global top
  top = []
```

```python
class Stack :
  def __init__(self) :    # 생성자
    self.top = []     # top이 클래스의 멤버변수가 됨

  def isEmpty(self) :   # 원래는 매개변수 없이 () 했지만, 클래스 안에서는 모든 메소드의 첫 파라미터로 self 추가 
    return len(self.top) == 0
  def size(self) :
    return len(self.top)
  def clear(self) :
    self.top = []   # 클래스 안에서니까 전역변수 선언 필요 X

  def push(self, item) :
    self.top.append(item)
 
  def pop(self) :   # 길이 있는지 확인
    if not self.isEmpty() :   # ()인 게 신기
      return self.top.pop(-1)   # 리스트인데 [-1] 아니라 (-1) 인가..

  def peek(self) :    # 리스트의 마지막 항목 반환
    if not self.isEmpty() :
      return self.top[-1]   # top은 리스트, 위에는 pop으로 끝나니까 (-1)? 


```

```python
odd = Stack()
even = Stack()

for i in range(10) :
  if i%2 == 0 :
    even.push(i)
  else :
    odd.push(i)
print('스택 even push 5회 :', even.top)
print('스택 odd push 5회 :', odd.top)
print('스택 even peek :', even.peek())
print('스택 odd peek :', odd.peek())

for s in range(2) :
  even.pop()
for s in range(3) :
  odd.pop()

print('스택 even pop 2회 :', even.top)
print('스택 odd pop 3회 :', odd.top)
# 아 이거 출력이 이상한게 아니네 ㅋㅋ 
# even은 스택 객체지만, even.top은 리스트 객체
```

```python
# 계산기의 후위표기 계산은 스택을 이용한다! 
# 과정을 분석, 그걸 알고리즘으로 표현, 그래서 계산 
# expr1 = ['8', '2', '/', '3', '-', '3', '2', '*', '+']  => 연산자, 피연산자 모두 문자열로 저장
# 스택 만들기/ 피연산자는 저장/ 연산자면 처음 꺼낸거 두번째로, 그 다음 꺼낸거 첫번째로 연산/ 연산 결과는 다시 스택에 저장/ 마지막 결과 하나 반환/ 피연산자는 float
def evalPostfix(expr):
  s = Stack()
  for elem in expr :
    if elem in ['+', '-', '*', '/']:   # 왜     
      val2 = s.pop()   # 어차피 피연산자가 먼저 나오겠고 바로 else로 가니까 스택에 저장되어 있겠네
      val1 = s.pop()   # expr에 있는걸 s(stack)에 저장
      # 계산을 해야한다.. 계산 
      if elem == '+':
        s.push(val1 + val2)    # 연산값은 스택에 넣는거 맞는데 피연산자는 스택에 넣는거 왜 없지?
      elif elem == '-':
        s.push(val1 - val2)
      elif elem == '*':
        s.push(val1 * val2)
      elif elem == '/':
        s.push(val1 / val2)
    else :    # 연산을 먼저 하든 피연산을 먼저 하든 상관없다/ 어차피 위에서 연산자들 아니면 바로 else로 넘어오겠네
      s.push(float(elem))    # float(), int() : 형 변환 함수 (Type Transpose)

  return s.pop()    # def 기준 들여쓰기 
```

```python
# 중위표기를 후위표기로 변환
# 연산자의 우선순위 : 1.괄호  2.곱나눔  3. 플마
def precedence(op):   # op는 연산자 전체를 뜻함
  ## 매우 중요/ if op == a or b 해버리면 애초에 무조건 트루가 되어버림 그래서 일일이 따로 코드 적어줘야 함.
  if op == '(' or op == ')' :   # 하긴 수식에서는 소괄호 하나네
    return 0      
  if op == '+' or op == '-' :
    return 1
  if op == '*' or op == '/' :
    return 2
  else :
    return -1   # 제일 마지막/ 숫자가 제일 우선 

def Infix2Postfix(expr):    # 중위수식 to 후위수식으로 
  s = Stack()
  output = []   # 연산자 모을 리스트
  for i in expr :
    if i == '(' :
      s.push(i)
    elif i == ')' :
      while not s.isEmpty():    # for과 while 차이는? while : 조건을 만족할 때 반복
        op = s.pop()    # 괄호 안에 연산자 다 꺼낸다
        if op == '(' :    # 괄호를 생각해줄 필요 없어. 다 꺼내면 괄호는 사라짐
          break;  # while의 break인가
        else :
          output.append(op)  # 붙인다가 먼저고 왼쪽 소괄호 나오면 그만둔다 인것 같은데            
    elif i in '+-*/' :
      while not s.isEmpty() :
        op = s.peek()   # 스택에서 모두? 꺼내 출력/ 일단 꺼내봄?
        if precedence(i) <= precedence(op):   # 우선순위가 같은 경우도 먼저 출력
          output.append(op)   # 추가/ output은 후위표기 리스트
          s.pop()   # 추가했으니 삭제?/ output이랑 순서 바뀌어야하는거 아닌가
        else : break
      s.push(i)   # 플마나 곱셈이 우선인 경우 스택에 저장되어 있는 중
    else :
      output.append(i)

  while not s.isEmpty():    # 비어있지 않다면
    output.append(s.pop())    # 모두 출력이란 뜻
  return output
```

```python
infix1 = ['8', '/', '2', '-', '3', '+', '(', '3', '*', '2', ')']
infix2 = ['1', '/', '2', '*', '4', '*', '(', '1', '/', '4', ')']
postfix1 = Infix2Postfix(infix1)
postfix2 = Infix2Postfix(infix2)
result1 = evalPostfix(postfix1)   # 아 내가 같은 i로 지정했구나
result2 = evalPostfix(postfix2)
print('중위표기 :', infix1)
print('후위표기 :', postfix1)
print('계산결과 :', result1, end='\n\n')
print('중위표기 :', infix2)
print('후위표기 :', postfix2)
print('계산결과 :', result2)
```

```python
# 미로 => 깊이 우선 탐색 (DFS)
map = [['1','1','1','1','1','1'],
       ['e','0','0','0','0','1'],
       ['1','0','1','0','1','1'],
       ['1','1','1','0','0','X'],
       ['1','1','1','0','1','1'],
       ['1','1','1','1','1','1']]
MAZE_SIZE = 6
# 칸의 위치는 (열,행) 순서 => (x,y) 좌표로 됨
def isValidPos(x,y):    # (x,y)가 갈 수 있는 방인지 검사
  if x<0 or y<0 or x>=MAZE_SIZE or y>=MAZE_SIZE :  # 키 반발이 좋군
    return False
  else :
    return map[y][x] == '0' or map[y][x] == 'X'   # map 안에서 갈 수 있는 곳은 방('0')과 출구('X')
```

```python
def DFS():
  stack = Stack()
  stack.push( (0,1) )   # 시작위치, 튜플() 자체를 삽입
  print('DFS :')

  while not stack.isEmpty() :   # 콜론 해주기 
    here = stack.pop()
    print(here, end=' => ')
    (x,y) = here
    
    if (map[y][x] == 'X') :   # 출구면
      return True
    else :
      map[y][x] = '.'     # 방문했음 체크/ map 위치에 문자를 추가할 수 있나? 
      if isValidPos(x, y-1) :   # 상    #(x,y-1)이 '0' or 'X' 이면 True가 나옴
        stack.push( (x, y-1) )    # 괄호 더블인거 체크
      if isValidPos(x, y+1) :
        stack.push( (x, y+1) ) 
      if isValidPos(x-1, y) :
        stack.push( (x-1, y) )  # 좌
      if isValidPos(x+1, y) :
        stack.push( (x+1, y) ) 
    print('현재 스택 :', stack)
  return False    # 스택이 비었다. while not 해놓은거 반대는 return으로 
  # 스택이 비어있지 않다 => 출구이거나 지금스택 출력하거나/ 스택이 비어있으면 False
```
