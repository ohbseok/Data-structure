```python
# 선형 큐에서는 삭제연산 0(n)이다/ 삭제 연산도 0(1)로 만들면 원형 큐를 이용하자
# front는 항상 공백 => front에 항목이 있으면 오류
# items가 MAX_QSIZE의 항목을 이미 가지고 있어야 함.

MAX_QSIZE = 10
class CircularQueue :
  def __init__(self):   # CircularQueue 생성자
    self.front = 0    # 같다가 아니라 위치
    self.rear = 0
    self.items = [None] * MAX_QSIZE   # 항목 저장용 리스트

  def isEmpty(self):
    return self.front == self.rear
  def isFull(self):
    return self.front == (self.rear+1) % MAX_QSIZE    # 나머지가 0이 될테니까 / 그냥 0, -1 하면 안되나? =>  이러면 선형이 되어버림, 나눠줘야 순환
    
  def clear(self):
    self.front == self.rear

  def enqueue(self, item):
    if not self.isFull() :    # 원형큐는 사이즈 정해져있어서 다 찼는지 확인/ 그러니 isFull도 했겠지/ 소괄호 무조건 해주기 => 안해주면 조건성립 안함
      self.rear = (self.rear+1) % MAX_QSIZE   # 왜 굳이 맥스사이즈로 나눴지 => 순환으로 만들어주려고
      self.items[self.rear] = item
    
  def dequeue(self):
    if not self.isEmpty():
      self.front = (self.front+1) % MAX_QSIZE   # 선형큐는 맨앞 항목 제거하면 뒤에게 순차적으로 앞으로 이동/ 원형큐는 항목 자리는 그대로 front가 한칸 뒤로
      return self.items[self.front]   # 이 부분 오타 왤케 많아 

  def peek(self):   # 스택에서는 맨 위, 큐에서는 맨 앞/ 출력위치는 정해져있으니 함수할 때 해 줬고, 여기서는 그냥 이용하면 됨
    if not self.isEmpty():
      return self.items[(self.front+1) % MAX_QSIZE]   # 맞네 front는 항상 비어져있고, 전단항목은 front 바로 뒤에 있으니/ 그리고 나누기 해줘야 사이클 반복이 되니까 

  def size(self):
    return (self.rear - self.front + MAX_QSIZE) % MAX_QSIZE   # rear가 front보다 더 작아질 수 있으니 항상 양수 만들어주고, 결과값에는 영향 없도록

  def display(self):    # front 다음 항복 - rear 항목까지
    out = []    # 출력할 빈 리스트가 필요하다?
    if self.front < self.rear :
      out = self.items[self.front+1 : self.rear+1]    # end 부분은 제외
    else :
      out = self.items[self.front+1 : MAX_QSIZE] + self.items[0 : self.rear+1]
      # ㅋㅋㅋㅋ 어쩐지 한 사이클 넘어가면 출력이 안되더니 ㅋㅋㅋ out 대신 output 시발 
      # 역순은 안됨/ 큐는 선입선출이라서 
      # front 칸은 공백이니 +1/ MAX_QSIZE까지 하면 최대 숫자까지 나옴
    print('[f=%s,r=%d] => ' %(self.front, self.rear), out )   # 함수 출력 다시
```

```python
# 덱
# 파이썬 모듈에서는 삭제 dequeue() 대신 get()
# 덱에서는 삭제 dequeue => delete/ 꺼내지 않고 반환 peek => get
# 큐에서는 한 방향 => 시계방향만 가능/ 덱에서는 반시계 회전도 가능  

class CircularDeque(CircularQueue):   # CircularQueue에서 상속/ 자식 안에 부모가 있다/ Deque안에 Que가 있다
# 부모 클래스에서 멤버변수와 메소드는 가져옴/ 생성자는 상속되지 않아서 다시 정의
  def __init__(self):
    super().__init__()    # 부모 클래스의 생성자를 호출

  # 부모 클래스의 멤버들은 super() 함수로 재사용 가능/ super().front = self.front 동일
  
  # 이미 큐에 있는 기능들 호출
  def addRear(self, item):
    self.enqueue(item)    # 더하는건 item을 넣어야하고 출력은 없음
  def deleteFront(self):
    return self.dequeue() # 빼는건 item 지정 X/ 그러나 반환 해야함/ 삭제된 항목을 반환  
  def getFront(self):
    return self.peek()
  
  # 이제는 덱에만 있는 기능들
  def addFront(self, item):   # 없는데 빼거나, 찼는데 더하면 오류남/ 무조건 확인하기!!
    if not self.isFull():
      self.items[self.front] = item   # 항목 저장 먼저
      self.front = self.front -1    # 이젠 -1한게 front에 입력
      if self.front < 0 : 
        self.front = MAX_QSIZE -1     # 왜 여기서는 %로 순환 안해주지?

  def deleteRear(self):
    if not self.isEmpty():
      item = self.items[self.rear]
      self.rear = self.rear -1
      if self.rear < 0 :
        self.rear = MAX_QSIZE -1    # 이거 이해 불가 
      return item

  def getRear(self) :   # 삭제하지 않고 반환만
    return self.items[self.rear]
    
```

```python
# 우선순위 큐 => 가장 일반적인 큐 Why? 우선순위 정하는 거에 따라 스택/큐 얼마든지 사용가능해서
# 파이썬 리스트를 이용한 우선순위 큐 ADT 구현

class PriorityQueue :
  def __init__(self):
    self.items = []

  def isEmpty(self):
    return len(self.items) == 0   # True/False인지 출력
  def size(self):
    return len(self.items)
  def clear(self):
    self.items = []

  # 이제는 삽입/ 리스트를 정렬하지 않으므로 리스트 맨 뒤에 추가하면 됨 Why?
  def enqueue(self, item):
    self.items.append(item)

  # 삭제 하려면 최우선순위 필요/ 찾는 메소드
  def findMaxIndex(self):
    if self.isEmpty():    # 빈 리스트가 아니라면(X)/ 빈 리스트면 None을 출력
      return None
    else :
      highest = 0   # 숫자 0으로 두는게 아니라 0번째 위치
      for i in range(1, self.size()):   # 여기에 콜론 안 붙이니 다음 문장도 들여쓰기 안되고/ size(self) 아님 
        if self.items[i] > self.items[highest]:
          highest = i
      return highest    # highest 위치의 인덱스를 반환함

  def dequeue(self):
    highest = self.findMaxIndex()   # 위에 메소드를 이용해서 최우선순위 지정/ self.메소드 하는거 맞네
    if highest != None :   # dequeue 생성하는 코드니까 dequeue 쓰면 안됨/ findMaxIndex : 비어있으면 None, 아니라면 highest 출력
    # is not/ != 결과값은 같네
      return self.items.pop(highest)    # 더하는건 반환 X/ 빼는건 삭제된 항목을 반환

  def peek(self):
    highest = self.findMaxIndex()   # highest는 findMaxIndex메소드로 찾은 인덱스이므로 메소드 뒤에 꼭 () 붙이기
    if highest != None :    # if문에 콜론 꼭 붙이기
      return self.items[highest]
```
