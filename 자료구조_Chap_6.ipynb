```python
# 연결된 구조
# 단순연결리스트 : 연결된 스택 (스택과 리스트를 구현해보자)
# 노드 클래스와 스택 클래스 각각 구현

class Node :
  def __init__(self, elem, link=None):   # 디폴트 인수 사용(link-None)
    self.data = elem
    self.link = link

# 또 다른 클래스
class LinkedStack :
  def __init__(self):
    self.top = None   # top 생성 및 초기화/ 이건 가리키고 있는 상태

  def isEmpty(self):
    return self.top == None   # 이건 같은 상태/ 정의할 땐 isEmpty(self), 메소드 쓸 때는 self.isEmpty()
  def clear(self):
    self.top = None   # 초기화는 가리킬 때

  def push(self, item):
    n = Node(item, self.top)    # item 데이터로 새로운 노드 n 지정 + n의 링크가 시작노드(self.top) 가리키도록
    self.top = n
  
  def pop(self):    # 노트북 100만원 아까워서 안 사고 뻐겼지. 그러다가 결국 대학원 들어가면서 사고 
    if not self.isEmpty():
      n = self.top
      self.top = n.link   # n.link는 사실 노드B임 (top이 다음노드를 가리키도록 지정)
      return n.data

  def peek(self):
    if not self.isEmpty():
      return self.top.data    # data를 반환하는거니까 top.data 해야 함

  def size(self):   # 비어있으면 0을 출력하니까 not isEmpty() 안해줘도 됨
    node = self.top   # 최초 n은 top (시작노드)
    count = 0
    while not node == None :
      node = node.link
      count += 1
    return count

  def display(self, msg='LickedStack'):
    print(msg, end='')
    node = self.top
    while not node == None :
      print(node.data, end='')
      node = node.link
    # print()
```

```python
# 단순연결리스트 : 연결된 리스트

class LinkedList :
  def __init__(self):
    self.head = None  # head 생성 및 초기화

  def isEmpty(self):    # 공백상태는 head가 None인지 검사
    return self.head == None    # 여기서는 가리키는게 아니라 왜 같다지?/ return 하면 True/False 결과값 나옴 
  def clear(self):
    self.head = None   # 초기화는 None 가리도록/ 가리키는건 출력 따로 없음

  def size(self):   # 비어있으면 0을 출력하니까 not isEmpty() 안해줘도 됨
    node = self.head   # 최초 n은 top (시작노드)
    count = 0
    while not node == None :
      node = node.link
      count += 1
    return count

  def display(self, msg='LickedLink'):
    print(msg, end=' ')
    node = self.head
    while not node == None :
      print(node.data, end=' ')
      node = node.link
    print()     # 이걸 해줘야 줄바꿈이 됨/ 안 하면 한 줄에 다 됨/ def에 대한 print

# 이제 리스트 ADT에 없는 연산

  # pos번째 node를 반환하는 연산
  def getNode(self, pos):
    if pos < 0 :
      return None
    else : 
      node = self.head
      while pos > 0 and node != None :    # pos=0이 되거나 node==None이 되면 종료
        node = node.link
        pos -= 1    # += -1 이랑 같겠지?
      return node   # node 자체 반환

  # pos번째 항목의 데이터만을 반환
  def getEntry(self, pos):    # 위에 함수 지정했으니 굳이 다시 할 필요 없네.. 함수를 쓰자
    node = self.getNode(pos)     # getNode(self, pos) <= 이거 아니다/ 지정한 함수 쓸 때는 self.getNode 로
    if node == None :   # 공백이라면! / 무조건 체크해줘야 함
      return None
    else :
      return node.data

  def replace(self, pos, elem):   # input 대신 튜플 안에서 elem 추가
    node = self.getNode(pos)    # 교체는 빼낼 필요 없다. 그냥 그 자리에 elem 이라고 지정해주면 자동으로 교체
    if node == None :   # if node != None/ 이거 하나로 대체할 수 있나?? 
      return None
    else :    # if 있다고 꼭 else 써야하는 건 아님 
      node.data = elem

  def find(self, data):
    node = self.head
    while node != None :    # 모든 노드에서 찾음/ while은 반복하다가 조건 만족하지 못하면 밑으로 가는게 아니라 그 위치에서 종료
      if node.data == data :
        return node
      node = node.link
  # ?   
    return None     # 찾아지지 않으면 None 반환/  while 만족 못하면 밑으로 내려오고 그때 def문에 대한 return으로 None

  def insert(self, pos, elem):
    before = self.getNode(pos-1)
    if before == None :   # 맨 앞인 경우 head 변경
      self.head = Node(elem, self.head)   # 원래 self.head 자리에 elem 놓고 다시 self.head 새로운게 그걸 참조하도록\
    else :
      node = Node(elem, before.link)    # pos 칸에 바로 노드N이 노드C를 가리키도록 before.link/ 이때까지는 before이 C를 가리키고 있다
      before.link = node    # 이제는 before이 node 가리키고 node는 pos 위치에

  def delete(self, pos):
    before = self.getNode(pos-1)
    if before == None :   # 시작이니까 시작을 삭제
      if self.head != None :    # 공백이 아니면 (head가 가리키는게 있으면)/ self.head = None => 공백이란 뜻
        self.head = self.head.link    # head를 다음으로 이동/ 우변이 먼저/ 그러면 원래 self.head는 삭제 
    elif before.link != None :    # else(before != None) 이거랑 다른가/ 
      before.link = before.link.link
    # if문 빠져나오는게 before이 있다는거고, before.link(pos 위치)가 있다면 다음다음으로 변경

# def한 것들 같은 코드셀에서 class에 종속시켜줘야 함
```

```python
# 원형연결리스트 : 연결된 큐
# 일반적인 큐는 선형/ 하지만 더 효율적으로 사용하기 위해 원형연결리스트
class CircularLinkedQueue :
  def __init__(self):
    self.tail = None    # tail : 유일한 데이터

  def isEmpty(self):
    return self.tail == None    # True/False로 나오니까 return
  def clear(self):
    self.tail = None   # 둘의 차이점을 정확히 알자/ 이건 초기화 지정만 (출력X)
  def peek(self):
    if not self.isEmpty():    # 항상 체크
      return self.tail.link.data    

  def enqueue(self, item):
    n = Node(item, None)    # 큐가 공백이든 아니든 수행해야 하니까 위로 빼기
    if self.isEmpty():    # 나머지에도 tail 다 적어야 한다
      n.link = n
      self.tail = n    # n 하나만 생성되었으니 이게 테일
    else :     # n 생성했고
      n.link = self.tail.link
      self.tail.link = n
      self.tail = n      

  def dequeue(self):    # 어차피 전단에서만 삭제 가능하니까 pos 없음
    if not self.isEmpty():
      n = self.tail.link    # n이 전단(무조건 self.tail로 기억)을 가리키도록/ 임의의 n으로 데이터를 반환하도록
      data = n.data    # 전단의 데이터
      if n.data == self.tail.data :   # 항목이 하나
        self.tail = None
      else :
        self.tail.link = n.link   # tail.link 옮겨주기       
      return data

  def size(self):   
    if self.isEmpty():    # 1부터 정의한거니 비어 있으면 0으로 해주기
      return 0
    else :
      n = self.tail.link
      count = 1
      while n != self.tail :    # while not n == sefl.tail 이랑 같나?
        n = n.link
        count += 1
      return count 
  def display(self, msg = 'CircularLinkedQueue :'):
    print(msg, end='')
    if not self.isEmpty():    # 꼭 공백인지 아닌지 체크
      n = self.tail.link    # size처럼 하나하나 들어가봐야 함/ 처음
      while n != self.tail :    # self.tail이 아니면 계속 반복
        print(n.data, end=' ')
        n = n.link
      print(n.data, end='')    # 루프 종료 조건에 주의! 처음부터 시작했으니 self.tail은 반복문 탈출 그러니 따로 지정
    print()     # 줄바꿈 표시임 
```
