```python
# car class 생성
class car:
  def __init__(self, color, speed=0):
    self.color = color
    self.speed = speed
```

```python
class car:
  def speedup(self):
    self.speed += 10
  def speeddown(self):
    self.speed -= 10
```

```python
def isEqual(self, carB):
  if self.color == carB.color :
    return True 
  else :
    return False
```

```python
# car class 참조해서 자식 클래스 생성

class supercar(car):
  def __init__(self, color, speed = 0, bTurbo = True):
    super().__init__(color, speed)
    # super() 라는 게 호출이란건가?
    # 이렇게 물음표 해놓은거 리스트로 정리해서 확실히 알아둬야 하는데... 그래야 내 것이 채워지는데
    self.bTurbo = bTurbo
```

```python
# 일단은 할거 하자 근데 보니까 ㅋㅋ 맞네 
# 자식 클래스에서 부모를 부르는게 super()
s1 = supercar("Gold", 0, True)
s2 = supercar('white', 0, False)
# 참조하는 형식임 
```

```python
def setTurbo(self,bTurbo = True):
  self.bTurbo = bTurbo
```

```python
def speedup(self):
  if self.bTurbo:
    self.speed += 50
  else : 
    super().speedup()
```

```python
def __str__(self):
  if self.bTurbo:
    return '[%s] [speed = %d] 터보모드' % (self.color, self.speed)
  else:
    return '[%s] [speed = %d] 일반모드' % (self.color, self.speed)
```

```python
s1.speedup()
s2.speedup()
print('슈퍼카1 :', s1)
print('슈퍼카2 :', s2)
```
